```python
# Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use
# MongoDB over SQL databases?

# MongoDB is a NoSQL, document-based database that stores data in JSON-like documents. Non-relational databases like MongoDB are suitable when dealing with unstructured or semi-structured data, providing more flexibility and scalability. MongoDB is preferred in scenarios requiring flexible schemas, real-time analytics, or high scalability, as it allows easier horizontal scaling and accommodates evolving data structures more efficiently than traditional SQL databases.

# Q2. State and Explain the features of MongoDB.

# MongoDB features include:
# - Flexible Schema: Documents can have varying structures.
# - Scalability: Horizontal scaling by distributing data across servers.
# - High Performance: Support for indexing and query optimization.
# - Replication & High Availability: Data redundancy for reliability.
# - Rich Query Language: Support for various query types.
# - GridFS: Storage of large files exceeding BSON document size limits.

# Q3. Write a code to connect MongoDB to Python. Also, create a database and a collection in MongoDB.

from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017/')
# Create a database named 'my_database'
my_db = client['my_database']
# Create a collection named 'my_collection' in 'my_database'
my_collection = my_db['my_collection']

# Q4. Using the database and the collection created in question number 3, write a code to insert one record,
# and insert many records. Use the find() and find_one() methods to print the inserted record.

# Insert one record
my_collection.insert_one({"name": "Alice", "age": 30})
# Insert multiple records
data_to_insert = [
    {"name": "Bob", "age": 25},
    {"name": "Charlie", "age": 35}
]
my_collection.insert_many(data_to_insert)

# Print inserted record using find_one()
print(my_collection.find_one({"name": "Alice"}))

# Print all inserted records using find()
for record in my_collection.find():
    print(record)

# Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to
# demonstrate this.

# The find() method is used to query documents in a collection based on specified criteria.
# Example:
for record in my_collection.find({"age": {"$gte": 30}}):  # Find records where age is greater than or equal to 30
    print(record)

# Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.

# The sort() method in MongoDB is used to arrange retrieved documents in a specified order (ascending or descending) based on a field.
# Example:
# Sorting documents based on 'age' field in ascending order
sorted_records = my_collection.find().sort("age", 1)
for record in sorted_records:
    print(record)

# Q7. Explain why delete_one(), delete_many(), and drop() is used.

# - delete_one(): Removes a single document that matches the specified criteria.
# - delete_many(): Removes multiple documents that match the specified criteria.
# - drop(): Drops an entire collection from the database.

# Example:
# Delete a document
my_collection.delete_one({"name": "Alice"})
# Delete multiple documents
my_collection.delete_many({"age": {"$gte": 30}})
# Drop the collection
my_collection.drop()
```
