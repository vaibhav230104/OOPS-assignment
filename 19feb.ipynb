```python
# Q1. What is Flask Framework? What are the advantages of Flask Framework?

# Flask is a micro web framework for Python, used to build web applications. It is minimalistic and lightweight, providing tools and libraries to create web applications quickly and easily. Advantages of Flask include:
# - Lightweight: Simple and easy to get started with.
# - Flexible: Offers freedom and flexibility in building web applications.
# - Extensible: Provides a wide range of extensions to add functionalities.
# - Integrated Development Server: Comes with a built-in development server.
# - Jinja2 Templating: Supports Jinja2 templating for flexible and dynamic content rendering.

# Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
# Jupyter Notebook.

from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run()

# For the screenshot, you can run this code in a Python environment and capture the output.

# Q3. What is App routing in Flask? Why do we use app routes?

# App routing in Flask refers to the process of mapping URLs to view functions. It's achieved using the @app.route decorator in Flask. We use app routes to define different URL endpoints and their corresponding functionalities or responses in the web application.

# Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
# route to show the following details:
# Company Name: ABC Corporation
# Location: India
# Contact Detail: 999-999-9999
# Attach the screenshot of the output in Jupyter Notebook.

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_details():
    return '''Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999'''

# Similar to Q2, to capture the output, you can run this code and take a screenshot of the displayed content.

# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
# url_for() function.

# The url_for() function in Flask is used for URL building. It generates URLs based on endpoint names.
# Example:
from flask import url_for

with app.test_request_context():
    print(url_for('hello_world'))
    print(url_for('welcome'))
    print(url_for('company_details'))

# This code, when run in a Flask application, will print the URLs associated with the provided endpoint names.
```
